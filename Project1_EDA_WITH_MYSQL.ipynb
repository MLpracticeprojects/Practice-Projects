### iv. CDF

### How is Overall perforamance of the class?

```python
counts, bin_edges = np.histogram(df['total'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.show();
```

Summary:<br>
1.We can see that  90% of students score above 500 marks
